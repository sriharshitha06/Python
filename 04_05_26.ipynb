```python
x = "AABBBCCCCDDDAAA"
res=""
c=1
for i in range(len(x)):
  if i+1<len(x) and x[i]==x[i+1]:
     c=c+1
  else:
    res=res+x[i]
    res=res+str(c)
    c=1
print(res)
```

```python
def fac(a):
  if a==1:
    return 1
  else :
    return a*fac(a-1)
a=6
x=fac(a)
print(x)
```

```python
def flower(a):
  if a==4:
    return
  print(a+10,end=" ")
  flower(a+1)
  print(a,end=" ")
flower(1)
```
